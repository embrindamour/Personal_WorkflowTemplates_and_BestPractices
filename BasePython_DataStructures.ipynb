# Base Python 

```python
# Lists
# adds an item
l = ['a','b','c','d']
l.append('e')
print(l)

# insert item f at index 1
l.insert(1,'f') 
print(l)

# adding mutliple items (unpacks a new list)
l2 = ['h', 'i']
l.extend(l2)
print(l)

# removes and then returns item at index 0
l.pop(0) # OUTPUT: item at index 0
print(l)

# Finds first occurrence of val and removes it (no return).
l.remove('c') 
print(l)

# deletes item at index 3
del l[3] 
print(l)

l[0] = 'z' 
print(l)
```

```python
# Dictionaries
user = {"name": "Alice", "age": 25}
new_data = {"age": 26, "city": "NY"}

# Merges other dict into this one. Overwrites existing keys
user.update(new_data)
print(user)

user['app'] = 'Snapchat'
print(user)

# Removes key and returns its value.
user.pop('age')
print(user)

# Removes & returns the last inserted key-value pair (tuple).
user.popitem()

# Returns value if exists, else returns None (safe access).
user.get('name')
```
